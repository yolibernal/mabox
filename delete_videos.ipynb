```python
import os

# https://stackoverflow.com/questions/7833715/python-deleting-certain-file-extensions
def remove_videos(path):
    for root, dirs, files in os.walk(path):
        for current_file in files:
            print ("Processing file: " + current_file)
            exts = ('.avi', '.mp4', ".AVI", ".mov", ".MOV", ".DS_Store")
            if current_file.lower().endswith(exts):
                os.remove(os.path.join(root, current_file))
```

```python
remove_videos("/Users/jonas/Documents/Familienbilder_no_videos")
```
